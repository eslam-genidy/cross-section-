```python
import googlemaps
import matplotlib.pyplot as plt
# define Google Elevation API key and client
api_key = 'AIzaSyAL46u8sFHbBR8nFroy1ycDC8Aii0VFtuE'
gmaps = googlemaps.Client(key=api_key)
# define two points as (latitude, longitude) coordinates
point1 = (21.422510,39.826168)
point2 = (24.470901,39.612236)
```

```python
# request elevation data from Google Elevation API
elevations = gmaps.elevation_along_path([point1, point2], samples=100)
# extract elevation values and distances from response
elevations2 = [result['elevation'] for result in elevations]
distances = [result['location']['lng'] for result in elevations]
```

```python
# plot the cross section
plt.plot(distances, elevations2)
plt.xlabel('Distance madina to meccah(m)')
plt.ylabel('Elevation (m)')
plt.show()
```
